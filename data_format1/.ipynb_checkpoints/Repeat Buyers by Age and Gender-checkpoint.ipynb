```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
merged = pd.read_csv('../data_format1/use_data/merged_dataset.csv')
merged
```

```python
user_log = pd.read_csv('../data_format1/use_data/train_format1[161-320].csv')
user_log.head()
```

```python
repeat = user_log.loc[user_log['label'] == 1]
repeat
```

```python
repeat_user_ids = repeat['user_id'].to_numpy()
repeat_user_ids
```

```python
user_repeat_transactions = merged.loc[merged['user_id'].isin(repeat_user_ids)]
user_repeat_transactions
```

```python
repeat_merchant_ids = repeat['merchant_id'].to_numpy()
repeat_merchant_ids
```

```python
merchant_repeat_transactions = merged.loc[merged['seller_id'].isin(repeat_merchant_ids)]
merchant_repeat_transactions
```
