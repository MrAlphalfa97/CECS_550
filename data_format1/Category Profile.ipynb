```python
#FEATURE 1: OVERALL ACTION COUNT/RATIO
actions = {}
for index, row in df.iterrows():
    if row['cat_id'] not in actions:
        # there may be multiple entries, so default with an empty list
        actions[row["cat_id"]] = [] 
    actions[row["cat_id"]].append(row["action_type"])
```

```python
actions
```

```python
action_count = {}
for key in actions.keys():
    action_count[key] = len(actions[key])
action_count
```

```python
total_action_count = sum(action_count.values())
total_action_count
```

```python
action_count_ratio = {}
for key in action_count:
    action_count_ratio[key] = action_count[key]/total_action_count
action_count_ratio
```

```python
df_cat['cat action count'] = -1
df_cat['cat action count ratio'] = -1
for index,row in df_cat.iterrows():
        df_cat.at[index, 'cat action count'] = action_count[row['cat_id']]
        df_cat.at[index, 'cat action count ratio'] = action_count_ratio[row['cat_id']]

df_cat
```

```python
#FEATURE 18: AGE RELATED FEATURES
cat_age = {}

# go through user_seller sub-dataframe and split entries by user_id:seller_id
for index, row in df.iterrows():
    if row['cat_id'] not in cat_age:
        # there may be multiple entries, so default with an empty list
        cat_age[row["cat_id"]] = [] 
    cat_age[row["cat_id"]].append(row["age_range"])

cat_age_avg = {}
for row in cat_age.items():
    cat_age_avg[row[0]] = np.nansum(row[1])/len(row[1])
sorted(cat_age_avg.items())
```

```python
total_cat_average_age = sum(cat_age_avg.values())/len(cat_age_avg.keys())
print("Average age total: ",total_cat_average_age)
```

```python
# For Other Columns, modify the groupby condition
male_counts = df.groupby(['cat_id'])['gender'].apply(lambda x: (x == 1).sum())
female_counts = df.groupby(['cat_id'])['gender'].apply(lambda x: (x == 0).sum())
unknown_gender_count = df.groupby(['cat_id'])['gender'].apply(lambda x: (x == 2).sum())
gender_total_item_counts = pd.DataFrame({'cat_male_count': male_counts, 'cat_female_count': female_counts, 'cat_unknown_gender_count': unknown_gender_count})
gender_total_item_counts
```

# Item Gender Count Ratio

```python
gender_counts = df_cat[['cat_male_count', 'cat_female_count', 'cat_unknown_gender_count']].sum(axis=1)
male_ratio = df_cat.apply(lambda row: row['cat_male_count'] / gender_counts[row.name], axis=1)
female_ratio = df_cat.apply(lambda row: row['cat_female_count'] / gender_counts[row.name], axis=1)
unknown_gender_ratio = df_cat.apply(lambda row: row['cat_unknown_gender_count'] / gender_counts[row.name], axis=1)
df_cat['item_male_count_ratio'] = male_ratio
df_cat['item_female_count_ratio'] = female_ratio
df_cat['item_unknown_count_ratio'] = unknown_gender_ratio
df_cat
```
