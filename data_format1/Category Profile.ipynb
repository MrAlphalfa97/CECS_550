```python
df = pd.read_csv("./use_data/expanded_training.csv")
df
```

```python
df_cat = df[['cat_id']]
df_cat
```

```python
#FEATURE 1: OVERALL ACTION COUNT/RATIO
actions = {}
for index, row in df.iterrows():
    if row['cat_id'] not in actions:
        # there may be multiple entries, so default with an empty list
        actions[row["cat_id"]] = [] 
    actions[row["cat_id"]].append(row["action_type"])
```

```python
actions
```

```python
action_count = {}
for key in actions.keys():
    action_count[key] = len(actions[key])
action_count
```

```python
total_action_count = sum(action_count.values())
total_action_count
```

```python
action_count_ratio = {}
for key in action_count:
    action_count_ratio[key] = action_count[key]/total_action_count
action_count_ratio
```

```python
df_cat['cat action count'] = -1
df_cat['cat action count ratio'] = -1
for index,row in df_cat.iterrows():
        df_cat.at[index, 'cat action count'] = action_count[row['cat_id']]
        df_cat.at[index, 'cat action count ratio'] = action_count_ratio[row['cat_id']]

df_cat
```

```python
#FEATURE 2: OVERALL DAY COUNT
days = {}

# go through user_seller sub-dataframe and split entries by cat_id:seller_id
for index, row in df.iterrows():
    if row['cat_id'] not in days:
        # there may be multiple entries, so default with an empty list
        days[row["cat_id"]] = [] 
    days[row["cat_id"]].append(row["time_stamp"])
```

```python
for k,v in days.items():
    days[k] = list(set(v)) # make each value list unique
days
```

```python
day_count = {}
for key in days.keys():
    day_count[key] = len(days[key])
day_count
```

```python
df_cat['cat day count'] = -1
for index,row in df_cat.iterrows():
        df_cat.at[index, 'cat day count'] = day_count[row['cat_id']]

df_cat
```
